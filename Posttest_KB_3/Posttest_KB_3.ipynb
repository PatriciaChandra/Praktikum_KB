## Import library yang diperlukan

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Masukkan file csv yang akan digunakan

```python
df = pd.read_csv('Top 2000 Universities of the World - Sheet1.csv')
df.head()
```

Mengubah nama colomn karena df tidak bekerja jika terdapat spasi pada parameter yang dituju

```python
df = df.rename(columns = {
    "World Rank" : "WorldRank",
    "National Rank" : "NationalRank",
    "Quality of Education Rank" : "Quality_of_Education_Rank",
    "Alumni Employment Rank" : "Alumni_Employment_Rank",
    "Quality of Faculty Rank" : "Quality_of_Faculty_Rank",
    "Research Performance Rank" : "Research_Performance_Rank",
})
df.head(7)
```

## Menampilkan grafik plot dan garis

```python
## Untuk memberikan judul serta label x dan y
plt.title('Grafik Perbandingan Score dengan World Rank')
plt.xlabel('World Rank')
plt.ylabel('Score')

## Untuk memunculkan grafik plot perbandingan Score dengan World Rank
plt.plot(df.WorldRank, df.Score)

## Berdasarkan grafik, world rank didapatkan berdasarkan besarnya score. Semakin besar atau bagus score yang didapat, maka semakin tinggi pula world rank yang didapat.
```

## Grafik Pie

```python
df = df.dropna(subset=['Score'])
```

```python
data_pie = df['Score'].value_counts().rename_axis('Total Score').reset_index(name='Score')
```

```python
plt.figure(figsize=(15,15))
plt.pie(data_pie.Score, labels=data_pie.Score, startangle=90,
autopct='%.1f%%')
plt.title('Total Score')
plt.show()
```

## Grafik Scatter Plot

```python
sns.scatterplot(data = df, x = "Score", y = "NationalRank")
plt.title('Grafik Score dan National Rank')
plt.show()

## Berdasarkan grafik yang ditampilkan, dapat dilihat bahwa national rank dapat dimiliki oleh beberapa score dan tidak hanya mementingkan besarnya score yang dimiliki
```

## Grafik Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Top 2000 Universities of the World - Sheet1.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.heatmap(df.corr(), center=0, cmap='Reds')
ax.set_title('Korelasi antar atribut pada data csv')

## Grafik ini dapat memperlihatkan korelasi antar variable
```
