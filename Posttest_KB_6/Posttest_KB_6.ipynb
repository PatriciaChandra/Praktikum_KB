## Hierarchical Clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets as dts
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv("heart.csv")
```

```python
h_cluster = AgglomerativeClustering(n_clusters=3)
```

```python
points = df[['age', 'sex', 'cp','trtbps']]
```

```python
agg_pred = h_cluster.fit_predict(points)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(points[agg_pred==0,0], points[agg_pred==0,1], c='red', label="cluster")
plt.legend()
plt.show()
```

## Dendrogram

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
merge = linkage(points,method='ward')
```

```python
plt.figure(figsize=(10,10))
dendrogram(merge,leaf_rotation=50,leaf_font_size=6)

plt.xlabel('Data Point')
plt.show()
```

```python
cluster_range = range(1,6)
inertias = []

for k in cluster_range:
    model = KMeans(n_clusters=k)
    model.fit(points)
    inertias.append(model.inertia_)
```

```python
plt.figure(figsize=(5,5))
plt.plot(cluster_range,inertias, '-o')
plt.xlabel('cluster ke - ')
plt.ylabel('inertias ')
plt.show()
```
